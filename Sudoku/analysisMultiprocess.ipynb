# Código para gerar os gráficos do tempo médio para resolver o Sudoku(3x3)

## importar as bibliotecas necessárias

```python
import sys
import os
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

## gerar uma lista com o nome de cada arquivo da Pasta Results 

```python
path_resl = './Results/'

for (dirpath, dirnames, filenames) in os.walk(path_resl):
    files = filenames
```

```python
rng = range(1,20)
rng2 = range(20,2510,20)
rng3 = range(1000,11000,1000)
rngt = rng + rng2 + rng3
list1_rest = ['time_{0}_process.csv'.format(nprocess) for nprocess in rng3]
```

## concatenar os arquivos em uma tabela de dados onde o número de processos é o nome da coluna e cada linha representa o tempo de duração do programa 

```python
list_df = [pd.read_csv(os.path.join(path_resl,ifiles)) for ifiles in list1_rest]
```

```python
data = pd.concat(list_df,axis=1)
data
```

```python
data_mean = data.apply(np.mean,axis=0)
data_std = data.apply(np.std,axis=0)
x_label = data.columns
```

## Função para gerar o gráfico de análise

```python

def graf_func(x=x_label,y=data_mean,err=data_std,size=(25,10)):
    plt.figure(figsize=size)
    fig = plt.errorbar(x,y,yerr=err)
    
    y_label = np.linspace(0,10,11)
    
    plt.grid()
    plt.xticks(x_label,rotation='vertical',fontweight='bold')
    plt.yticks(y_label,map(lambda x:str(round(x,2))+'s',y_label),
               rotation='vertical',fontweight='bold',fontsize=10)
    plt.xlabel('numero de processos',fontsize=20,fontweight='bold')
    plt.ylabel('tempo(s)',fontsize=20,fontweight='bold')
    
    plt.title(u'Tempo de processamento do Sudoku(3x3) para encontrar uma solução',fontweight='bold',fontsize=20)
    
    return fig,plt
```

## gerando o gráfico

```python
fig, plt_g = graf_func()
plt_g.savefig('./Results/Figures/temp_3x3_1000_10000_1000.png')
```
